## 🕒 Data Preprocessing: Date/Time Feature Extraction

The Date/Time column was initially stored as an object data type.

To perform time-based analysis, it was converted into a datetime format.

New features were extracted from the Date/Time column:

- **Year:** Identifies the year of observation
- **Month:** Helps analyze seasonal patterns
- **Day:** Allows daily-level analysis
- **Hour:** Helps study changes throughout the day

These extracted features will be used for time-based analysis and further feature engineering.

```python
data["Date/Time"]=pd.to_datetime(data["Date/Time"])
```

## 🌡️ Temperature Category Analysis

## 🌍 Seasonal Feature Creation

The month feature was transformed into seasonal categories using a custom function.

The months were grouped into four seasons:
- Winter: December, January, February
- Spring: March, April, May
- Summer: June, July, August
- Fall: September, October, November

This new feature helps analyze weather patterns across different seasons.

```python
def get_season(month):
    if month in [12,1,2]:
        return "Winter"
    elif month in [3,4,5]:
        return "Spring"
    elif month in [6,7,8]:
        return "Summer"
    else:
        return "Fall"
data["Seasons"]=data["Month"].apply(get_season)
```

```python
data["Seasons"].value_counts()
```

## 💧 Humidity Impact on Visibility

The data was grouped based on humidity categories to analyze how humidity levels affect visibility.

The average visibility was calculated for each humidity category to understand whether changes in humidity are associated with changes in visibility conditions.

```python
data["Humidity_Category"]=pd.cut(data['Rel Hum_%'],bins=[0,40,60,80,100],labels=["Low","Moderate","High","Very High"])
plt.figure(figsize=(8,5))

data.groupby("Humidity_Category")["Visibility_km"].mean().plot(
    kind="bar",
    color="Red"
)

plt.title("Average Visibility Across Humidity Levels")
plt.xlabel("Humidity Category")
plt.ylabel("Average Visibility (km)")

plt.xticks(rotation=0)
plt.show()
```

### Observation

The results indicate that visibility decreases as humidity increases.

Very high humidity conditions recorded lower average visibility, suggesting that moisture in the air can reduce visibility.

## 🔥 Feature Correlation Analysis

A correlation heatmap was created to analyze the relationships between numerical weather variables.

This helps identify whether variables have positive, negative, or weak relationships with each other.

The analysis can reveal patterns such as how temperature, humidity, visibility, wind speed, and pressure are related.

```python
import seaborn as sns
correlation=data[["Temp_C","Dew Point Temp_C","Rel Hum_%","Wind Speed_km/h","Visibility_km","Press_kPa"]].corr()
sns.heatmap(correlation,annot=True)
```

### Observation
- Correlation analysis revealed a very strong positive relationship(0.93) between temperature and dew point temperature, indicating that both variables  tend to to increase together.
- Relative Humidity showed a strong negative correlation(-0.63) with visibility, suggesting higher humidity levels are associated with reduced visibility.
- Most other variables exhibited weak to moderate correlations, indicating limited linear relationships between them

## 🌡️ Relationship Between Temperature and Dew Point

A regression plot was used to analyze the relationship between temperature and dew point.

The plot shows the distribution of data points along with a fitted linear trend line to understand whether dew point changes with temperature.

```python
sample_data = data.sample(300, random_state=42)
plt.figure(figsize=(19,5))

sns.regplot(
    x="Temp_C",
    y="Dew Point Temp_C",
    data=sample_data
)

plt.title("Temperature vs Dew Point Temperature")
plt.xlabel("Temperature (°C)")
plt.ylabel("Dew Point Temperature (°C)")

plt.show()
```

### Observation

The regression plot shows a positive linear relationship between temperature and dew point.

As temperature increases, dew point generally tends to increase as well, indicating that warmer conditions are often associated with higher moisture levels in the air.
